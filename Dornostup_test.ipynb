Вычисление среднего, максимального/минимального значений, медианы и моды для всего массива в целом. Это можно сделать разными способами. Можно вывести в одной таблице все показатели для всех переменных.

```python
data.info()
#видно, что есть пропущенные значения в некоторых переменных (фичей, столбцов)
```

```python
data.describe(exclude = 'object')
#описательная статистика только для численных переменных (фичей, столбцов)
```

А можно смотреть отдельно по каждой метрике для всех переменных. В Python есть встроенные функции mean, max, min, но они проигрывают по времени работы, функциям из библиотеки NumPy.

```python
np.mean(data)
```

```python
np.max(data)
#np.min(data)
```

```python
data.median()
```

```python
data.mode().head(5)
#для некоторых переменных получилось несколько мод, т.е. несколько дискретных значений, которые встречаются в выборке одинаковое количество (максимальное) раз.
```

Описательная статистика для каждого типа контента (столбец Type) в отдельности.

```python
data.groupby('Type').mean()
#data.groupby('Type').median()
#data.groupby('Type').min()
#data.groupby('Type').max()
```

Можно сразу все метрики по всем переменным, но на это неудобно смотреть.

```python
data.groupby('Type').agg(['mean', 'min', 'max', 'median'])
```

Можно посмотреть по каким-то отдельным переменным (например, share и likes).

```python
data.groupby('Type')[['share', 'like']].describe()
#посты с видео набирают, в среднем, больше лайков и репостов, чем остальные
```

Найти самый популярный объект в выборке. Не совсем понятно, что считать критерием популярности поста. Предположим, что это суммарное число лайков, комментариев и репостов (переменная Total Interactions).
